<h2 style = "font-size: 20px; font-weight: bold; text-align: left;">Importing Required Libraries</h2>



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

<h2 style = "font-size: 20px; font-weight: bold; text-align: left;">Loading the Dataset</h2>

```python
data =  pd.read_csv("USA_housing.csv")
data.head()
```

<h2 style = "font-size: 20px; font-weight: bold; text-align: left;">Drop address column</h2>

```python
data = data.drop(['Address'], axis=1)
data.head()
```

<h2 style = "font-size: 20px; font-weight: bold; text-align: left;">Check for Missing Data</h2>

```python
sns.heatmap(data.isnull())
```

<h2 style = "font-size: 20px; font-weight: bold; text-align: left;">Train Test Split</h2>

```python
X = data.drop('Price', axis = 1)
Y = data['Price']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .30);
```

<h2 style = "font-size: 20px; font-weight: bold; text-align: left;">Training and predicting</h2>

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

<h2 style = "font-size: 20px; font-weight: bold; text-align: left;">Evaluation</h2>

```python
error = np.sqrt(metrics.mean_absolute_error(Y_test, predictions))
```
